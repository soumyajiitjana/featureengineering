1. What is a parameter?

A parameter is a numerical value that describes a population.
Example: Average height of all students in a school (true value).

2. What is correlation?

Correlation is a statistical measure that describes the strength and direction of a relationship between two variables.

3. What does negative correlation mean?

Negative correlation means when one variable increases, the other decreases.
Example: Speed ↑ → Time taken ↓.

5. How does loss value help in determining whether the model is good or not?

The loss value shows how far the model’s predictions are from the actual values.

Low loss → Good model

High loss → Poor model

6. What are continuous and categorical variables?

Continuous variables: Numerical values that can take any number.
Example: height, weight, price.

Categorical variables: Values representing categories or labels.
Example: gender, city, product type.

17. What is causation? Explain difference between correlation and causation with an example.

Causation means one variable directly causes change in another.

Difference:

Correlation	Causation
Two variables move together	One variable causes the change
No proof of cause-effect	Direct cause-effect relationship

Example:
Ice cream sales ↑ and drowning cases ↑ (correlation) → both happen in summer, not causation.

18. What is an Optimizer? What are different types of optimizers? Explain each with an example.

An optimizer reduces loss by updating model parameters.

Common optimizers:

Gradient Descent
Moves down the slope of loss function.

Stochastic Gradient Descent (SGD)
Updates using 1 random sample at a time → faster.

Momentum
Adds speed to GD for faster convergence.

Adam
Combines Momentum + RMSProp
Most used optimizer.

23. What is feature scaling? How does it help in Machine Learning?

Feature scaling transforms variables so they lie on a similar scale.
It helps machine learning models:

Converge faster

Reduce bias due to large values

Improve accuracy

24. How do we perform scaling in Python?

Using StandardScaler:

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


Using MinMaxScaler:

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
